```python
# 막대그래프(bar graph)를 그린다.
means = [1, 2, 3]
stddevs = [0.2, 0.4, 0.5]
bar_labels = ['bar 1', 'bar 2', 'bar 3']

# plot bars
x_pos = list(range(len(bar_labels)))
plt.bar(x_pos, means, yerr=stddevs)

plt.show()
```

```python
# 히스토그램 (histogram)
rng = np.random.RandomState(123)
x = rng.normal(0, 20, 1000) 

# fixed bin size
bins = np.arange(-100, 100, 5) # fixed bin size

plt.hist(x, bins=bins)
plt.show()
```

```python
# 중첩 히스토그램
rng = np.random.RandomState(123)
x1 = rng.normal(0, 20, 1000) 
x2 = rng.normal(15, 10, 1000)

# fixed bin size
bins = np.arange(-100, 100, 5) # fixed bin size

plt.hist(x1, bins=bins, alpha=0.5)
plt.hist(x2, bins=bins, alpha=0.5)
plt.show()
```

### legend 표시

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), label=('sin(x)'))
plt.plot(x, np.cos(x), label=('cos(x)'))

plt.ylabel('f(x)')
plt.xlabel('x')

plt.legend(loc='lower left')
plt.show()
```

### subplot

```python
# import numpy as np
# import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

def g(t):
    return np.sin(np.pi*t)

t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.01)

plt.subplot(221)
plt.plot(t1, f(t1))

plt.subplot(222)
plt.plot(t2, g(t2))

plt.subplot(223)
plt.plot(t1, f(t1), 'r-')

plt.subplot(224)
plt.plot(t2, g(t2), 'r-')

plt.show()
```

```python
plt.subplot(211)
plt.plot(t1, f(t1))

plt.subplot(212)
plt.plot(t2, g(t2))

plt.show()
```

```python
# figure object, axis object
x = range(11)
y = range(11)  #(11,0,-1)

fig, ax = plt.subplots(nrows=2, ncols=3,
                       sharex=False, sharey=False)  # change sharex, sharey option

for row in ax:
    for col in row:
        col.plot(x, y)
        
plt.show()
```

### 그래픽 응용: 1차원 회기(regression)

```python
# 그래프 데이터 (x, y)
x = np.array([1.628, 3.363, 5.145, 7.683, 9.855])
y = np.array([1.257, 3.672, 5.841, 7.951, 9.775])

# 그래프를 그림
plt.plot(x, y, 'o', ms=12)   # 데이터를 그림
```

```python
# 최소 자승법(least square method)으로 데이터를 근사하는 직선을 구함
a = np.array([x, np.ones(x.size)])
a = a.T
m, c = np.linalg.lstsq(a, y)[0]  # 일차원 선형회기 계수

# 그래프를 겹쳐서 그림
# 데이터와 선형회기 근사직선을 동시에 그림.
t = plt.plot(x, y, 'o', ms=12)   # 데이터를 그림
t = plt.plot(x, (m*x+c))  # 근사직선을 그림, m:기울기,c:bias(절편)
```

#### 겹침 그림을 그리고 저장하기

```python
#plt.plot( x, y, 'o', ms=12,x, (m*x+c))   # positional argument 에러 디폴트 값을 선언한것은 항상 맨뒤로 가야함
plt.plot(x, (m*x+c), x, y, 'o', ms=12)   # 데이터와 근사직선을 동시에 그림
plt.savefig('linear_regression.png', dpi=75)
```
